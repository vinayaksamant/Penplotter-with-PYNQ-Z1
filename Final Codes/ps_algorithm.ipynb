```python
from PIL import Image
import random


class PSAlgorithm: 

    def __init__(self) -> None:
        pass

    def process_image(self, image: Image) -> tuple: 
        width, height, pix_data = self.___get_pixel_data(image)
        lines = self.__post_lines(config, pix_data)
        return (width, height, lines)

    def __calculate_brightness(self,pixel):
        # For RGB images, calculate luminance using the formula: 0.299*R + 0.587*G + 0.114*B
        if len(pixel) == 3:  # RGB image
            return 0.299 * pixel[0] + 0.587 * pixel[1] + 0.114 * pixel[2]
        elif len(pixel) == 1:  # Grayscale image
            return pixel[0]
        elif len(pixel) == 4:  # RGBA image (Background removed image)
            return (0.299 * pixel[0] + 0.587 * pixel[1] + 0.114 * pixel[2]) * (pixel[3] / 255.0)
        else:
            raise ValueError("Unsupported pixel format")

    def ___get_pixel_data(self, image: Image) -> tuple:
        # Get image dimensions
        width, height = image.size
        print("Image Dimension: ")
        print(width , height)

        # Get pixel data
        pixel_data = list(image.getdata())

        # Calculate brightness for each pixel
        brightness_list = [[self.__calculate_brightness(pixel_data[y * width + x]) for x in range(width)] for y in range(height)]
    
        return (width, height, brightness_list)
    
    def __post_lines(self,config : dict, pix_data: list) -> list:
        points = []
        spacing = config['Resolution']

        dot_rand = random.Random(str(config['Seed']))
        direction_rand = random.Random(str(config['Seed']))

        if config['Line Direction'] < 90:
            x_offset = config['Line Direction'] / 90.0 * spacing
            y_offset = spacing
        else:
            x_offset = spacing
            y_offset = spacing - (config['Line Direction'] - 90) / 90.0 * spacing

        y = 0
        while y < config['height'] - spacing:
            x = 0
            while x <= config['width'] - spacing:
                pixel_val = pix_data[y][x] if 0 <= x < config['width'] and 0 <= y < config['height'] else 0
                
                # p = pixel_val / 255.0
                # p = p * p * p * 0.5
                # p = p * spacing

                p = pixel_val / 255.0
                p = p  * 0.5
                p = p * spacing

                if p < dot_rand.random():
                    if config['Random Direction']:
                        direction = direction_rand.random()
                        if direction_rand.random() > 0.5:
                            points.append([[x + direction * spacing, y], [x + spacing - direction * spacing, y + spacing]])
                        else:
                            points.append([[x, y + direction * spacing], [x + spacing, y + spacing - direction * spacing]])
                    else:
                        points.append([[x + x_offset, y + y_offset], [x + spacing - x_offset, y + spacing - y_offset]])

                x += spacing

            y += spacing

        return points

# Example usage
config = {
    'Resolution': 2,
    'Line Direction': 0,
    'Random Direction': False,
    'Seed': 50,
    # 'width': 912,
    # 'height': 925
    'width': 640,
    'height': 360
}
```
